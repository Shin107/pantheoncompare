```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from astropy.coordinates import SkyCoord
import astropy.coordinates as coord
import astropy.units as u
plt.style.use(['science','notebook','grid'])
from astroquery.simbad import Simbad
JLA=pd.read_csv("jla_lcparams.txt",sep=" ")
PAN=pd.read_csv("Pantheon+SH0ES.dat",sep=" ")
pd.set_option('max_columns',None)
```

I try to search for thoese supernovae whihc are in JLA and Pantheon+ both,since for some of the supernovae the names of the supernovae are different in both the surveys hence I cant search for same name instead I use the condition of difference in PEAKMJD and the agular seperation.

## Condition 1
The difference in peakmjd < 3 days and angular separation $< 0.05 ^{\circ}$




```python
rd1=SkyCoord(JLA['ra'], JLA['dec'], unit=u.deg)
rd2=SkyCoord(PAN['RA'], PAN['DEC'], unit=u.deg)
gal1 = rd1.galactic
gal2 = rd2.galactic
ind1=np.array([])
ind2=np.array([])

jx=JLA['x1']
px=PAN['x1']

jz=JLA['zhel']
pz=PAN['zHEL']

JLAmjd=JLA['tmax']
PANmjd=PAN['PKMJD']
print('#,JLA[name],PAN[name],JLA[zcmb],PAN[zCMB],MJDJLA,MJDPAN')
s1=np.array([])
s2=np.array([])
for i in range (len(JLA['ra'])):
    for j in range (len(PAN['RA'])):
        
         if  coord.angular_separation(gal1[i].l,gal1[i].b,gal2[j].l,gal2[j].b)<0.05*u.deg and abs(JLAmjd[i]-PANmjd[j])<3:
            ind1=np.append(ind1,i)
            ind2=np.append(ind2,j)
            print(i,JLA['#name'][i],PAN['CID'][j],JLA['zcmb'][i],PAN['zCMB'][j],JLA['tmax'][i],PAN['PKMJD'][j])
            #s1=np.append(s1,JLA['zhel'][i]-PAN['zHEL'][j])
            #s2=np.append(s2,JLA['zcmb'][i]-PAN['zCMB'][j])

      
```

### Checking if the supernovae ids are same for all sne(i.e if they are same supernovae from simbad catalogue)

```python
same=0
SDSS=0
sn=0
sd=np.array([])
sd2=np.array([])
count=0
jlacc=JLA.iloc[indjla]
pancc=PAN.iloc[indpan]
inde=np.array([])
other=np.array([])
xx=np.array([])
other1=np.array([])
print("the only SNE with different ID's are ")

#due to different nomenclature some supernovae in pantheon+ have different names from JLA so here I check for those ID's
for indexx in range(len(jlacc['#name'])):
    i=jlacc['#name'].iloc[indexx]
    j=pancc['CID'].iloc[indexx]

    if i.lower()==j.lower():
        same+=1
        inde=np.append(inde,indexx)
    elif i[:4].upper()=='SDSS':
        if i[4:].lower()==j.lower():
            SDSS+=1
            inde=np.append(inde,indexx)
            
        else:
            print(i,j)
            sd=np.append(sd,i[4:])
            sd2=np.append(sd2,j.lower())
            if (Simbad.query_object("SDSS-II SN "+str(i[4:]))['MAIN_ID'][0][3:]==j.lower()):
                inde=np.append(inde,indexx)
            count+=1
    elif i[:2].lower()=='sn':
        if i[2:].lower()==j.lower():
            sn+=1
            inde=np.append(inde,indexx)
        else:
            print(i,j)
            if i[2:]=='2005ir':
                inde=np.append(inde,indexx)
            xx=np.append(xx,i[2:])
            count+=1
    else:
        print('---------------')
        print(i,j)
        
        other=np.append(other,i)
        other1=np.append(other1,j)
        count+=1
        
```
